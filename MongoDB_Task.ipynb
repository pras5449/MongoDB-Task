**1) Find the student name who scored maximum scores in all (exam, quiz and homework)?**

```python
list_all_marks=[]
for i in x.find():
  total=0
  for j in range(len(i['scores'])):
    total=total+i['scores'][j]['score']
  list_all_marks.append(total)  
  x.update_one({'_id': i['_id']},{'$set': {'total': total}})
```

```python
for i in x.find():
  if(i['total']==max(list_all_marks)):
    print(i)
    break
```

**3.Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.**

```python
for i in x.find():
  if (i['scores'][0]['score']>=40 and i['scores'][1]['score']>=40 and i['scores'][2]['score']>=40):
    x.update_one({'_id': i['_id']},{'$set': {'grade': 'pass'}})
  else:
    x.update_one({'_id': i['_id']},{'$set': {'grade': 'fail'}})
```
